## Visualization summary

- We normalized wage datasets for 2021–2024 to a consistent schema and annualized hourly median wages using 2080 hours where `annual_wage_flag == 0`.
- For each year, we grouped records by `noc_title_eng` and computed the maximum median annualized wage per job title.
- We selected the top 50 titles and rendered a horizontal bar chart using seaborn (`viridis` palette), sorted ascending for readability.
- Chart title: “Top 50 Highest Median Wages by Job Title in {year}”; axes labeled “Median Wage (Annualized)” and “Job Title”.

### What it shows
- Highest-compensated occupations each year, highlighting medical specialists, judges, senior managers, and other top-earning roles.
- Cross-year comparison is possible via the repeated charts for 2021–2024, giving a quick sense of consistency and change among top earners.

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

#load data
data_2021 = pd.read_csv('data/wages_2021.csv')
data_2022 = pd.read_csv('data/wages_2022.csv')
data_2023 = pd.read_csv('data/wages_2023.csv')
data_2024 = pd.read_csv('data/wages_2024.csv')

#load NOC elements data
noc_codes = pd.read_csv('data/noc_2021_version_1.0_-_elements.csv')

# Normalization of columns across years
def normalize(df, year):
    df = df.copy()
    df['year'] = year

# Rename columns to standard names
    column_map = {
        'NOC_Title': 'noc_title_eng',
        'NOC_TITLE_ENG': 'noc_title_eng',
        'NOC_Title_eng': 'noc_title_eng',

        'PROV': 'province',
        'prov': 'province',

        'ER_Name_Nom_RE': 'er_name_eng',
        'ER_Name': 'er_name_eng',

        'ER_Code_Code_RE': 'er_code',

        'Low_Wage_Salaire_Minium': 'low_wage',
        'Median_Wage_Salaire_Median': 'median_wage',
        'High_Wage_Salaire_Maximal': 'high_wage',
        'Average_Wage_Salaire_Moyen': 'avg_wage',

        'Quartile1_Wage_Salaire_Quartile1': 'q1_wage',
        'Quartile3_Wage_Salaire_Quartile3': 'q3_wage',
        'Non_WageBen_pct': 'non_wage_benefit',

        'Data_Source_E': 'data_source_eng',
        'Wage_Comment_E': 'wage_comment_eng',

        'Reference_Period': 'reference_period',
        'Revision_Date_Date_revision': 'revision_date',

        'Annual_Wage_Flag_Salaire_annuel': 'annual_wage_flag',
        'NOC_CNP': 'noc_code'
    }

    df = df.rename(columns=column_map)

    wanted_cols = [
        'year', 'noc_code', 'noc_title_eng', 'province',
        'er_code', 'er_name_eng', 'low_wage', 'median_wage',
        'high_wage', 'avg_wage', 'q1_wage', 'q3_wage',
        'non_wage_benefit', 'data_source_eng', 'reference_period',
        'revision_date', 'annual_wage_flag', 'wage_comment_eng'
    ]

    # Ensure all wanted columns exist
    for col in wanted_cols:
        if col not in df.columns:
            df[col] = np.nan

    for col in wanted_cols:
        if col not in df.columns:
            df[col] = np.nan

    df = df[wanted_cols]

# Clean NOC codes by removing "NOC_" prefix if present
    df['noc_clean'] = df['noc_code'].str.replace("NOC_", "", regex=False)


# Annualize wages based on annual_wage_flag
    df['median_wage_annualized'] = np.where(
        df['annual_wage_flag'] == 0, # hourly wage flag, if 0 then hourly 
        df['median_wage'] * 2080, # convert to annual assuming 2080 work hours
        df['median_wage']
    )

    return df


# ------------------------------
# APPLY NORMALIZATION TO ALL YEARS
# ------------------------------
norm_2021 = normalize(data_2021, 2021)
norm_2022 = normalize(data_2022, 2022)
norm_2023 = normalize(data_2023, 2023)
norm_2024 = normalize(data_2024, 2024)


# This function analyzes a given year's normalized dataframe and produces results and plots
def analyze_year(norm_df, year, top_n=50, show_plot=True):
    result = (
        norm_df
        .groupby('noc_title_eng', as_index=False)['median_wage_annualized']
        .max()
        .sort_values(by='median_wage_annualized', ascending=False)
    )
    top_n_df = result.head(top_n)

    print(f"Year {year} result (top {top_n} shown):")
    print(top_n_df)

    if show_plot and not top_n_df.empty:
        plt.figure(figsize=(16, 20))
        sns.set_style("whitegrid")
        sns.barplot(
            data=top_n_df.sort_values('median_wage_annualized', ascending=True),
            x='median_wage_annualized',
            y='noc_title_eng',
            palette='viridis'
        )
        plt.title(f'Top {top_n} Highest Median Wages by Job Title in {year}', fontsize=16, pad=20)
        plt.xlabel('Median Wage (Annualized)', fontsize=12)
        plt.ylabel('Job Title', fontsize=12)
        plt.tick_params(axis='y', labelsize=10)
        plt.tick_params(axis='x', labelsize=10)
        plt.gca().margins(y=0.02)
        plt.tight_layout()
        plt.show()

    return result, top_n_df

norms_by_year = {
    2021: norm_2021,
    2022: norm_2022,
    2023: norm_2023,
    2024: norm_2024,
}

results = {}
tops = {}

for y, df in norms_by_year.items():
    res, top = analyze_year(df, y, top_n=50, show_plot=True)
    results[y] = res
    tops[y] = top

```

## Identifying the most “AI‑driven” fluctuating jobs

Goal:
- Find occupations with the largest wage volatility across 2021–2024 and a clear upward trend that plausibly aligns with AI adoption.

Approach:
- Use normalized datasets `norm_2021`, `norm_2022`, `norm_2023`, `norm_2024`.
- Aggregate by occupation (prefer `noc_clean` or `noc_title_eng`) and province if needed.
- Compute:
    - Years present (count of years with non‑null `median_wage_annualized`).
    - Wage count (number of valid observations).
    - Volatility metrics: standard deviation and range (max–min) of `median_wage_annualized`.
    - Trend metric: slope of a simple linear fit (wage ~ year) or CAGR between first and last available year.
- Filter to occupations with:
    - At least 2–3 years of valid data.
    - Positive slope (rising wages).
    - High volatility (top decile by std or range).
- Rank and report top fluctuators, adding qualitative interpretation (e.g., AI exposure in job tasks, industry digitization).

Notes:
- `fluctuation` currently empty; we will populate it by joining the four yearframes and computing metrics above.
- Consider stratifying by industry or province to avoid geographic outliers.
- Optional: enrich with `noc_codes` to map `noc_clean` to NOC class titles for readable outputs.

```python
# Compute wage trends over years

all_years = pd.concat([norm_2021, norm_2022, norm_2023, norm_2024], ignore_index=True)
trend = (
    all_years.groupby(['noc_title_eng', 'year'])['median_wage_annualized']
             .median()
             .unstack()
)
trend['pct_change'] = (trend[2024] - trend[2021]) / trend[2021]

all_years['AI_Wage_Signal'] = all_years['noc_clean'].map(
    trend['pct_change'] < -0.05
)

print("Wage Trend Analysis (2021-2024):")
print(trend.sort_values('pct_change').head(20))

#display the data above on a plot percentage change over years
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
trend_sorted = trend.sort_values('pct_change').head(20)
sns.barplot(
    data=trend_sorted.reset_index(),
    x='pct_change',
    y='noc_title_eng',
    palette='coolwarm'
)
plt.title('Top 20 Job Titles with Largest Decrease in Median Wage (2021-2024)', fontsize=16, pad=20)
plt.xlabel('Percentage Change in Median Wage', fontsize=12)
plt.ylabel('Job Title', fontsize=12)
plt.tick_params(axis='y', labelsize=10)
plt.tick_params(axis='x', labelsize=10)
plt.gca().margins(y=0.02)
plt.tight_layout()
plt.show()

# Display the highest increase in median wage and plot
print("Top 20 Job Titles with Largest Increase in Median Wage (2021-2024):")
print(trend.sort_values('pct_change', ascending=False).head(20))
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
trend_sorted_inc = trend.sort_values('pct_change', ascending=False).head(20)
sns.barplot(
    data=trend_sorted_inc.reset_index(),
    x='pct_change',
    y='noc_title_eng',
    palette='viridis'
)
plt.title('Top 20 Job Titles with Largest Increase in Median Wage (2021-2024)', fontsize=16, pad=20)
plt.xlabel('Percentage Change in Median Wage', fontsize=12)
plt.ylabel('Job Title', fontsize=12)
plt.tick_params(axis='y', labelsize=10)
plt.tick_params(axis='x', labelsize=10)
plt.gca().margins(y=0.02)
plt.tight_layout()
plt.show()
```
